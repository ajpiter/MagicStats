Basketball-Reference data for Orlando Magic Player Chuma Okeke

Based on the Towards Data Science Article, 
Web Scraping NBA Stats, Sanchez 2019
https://towardsdatascience.com/web-scraping-nba-stats-4b4f8c525994

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

# NBA season we will be analyzing
year = 2021
# URL page we will scraping (see image above)
url = "https://www.basketball-reference.com/players/o/okekech01/gamelog/2021/".format(year)
# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html, features="html.parser")

headers = ["G", "Date", "Age", "Tm", "Location", "Opp", "W/L", "GS", "MP", "FG", "FGA", "FG%",
           "3P", "3PA", "3P%", "FT", "FTA", "FT%", "ORB", "DRB", "TRB", "AST", "STL", "BLK", "TOV",
           "PF", "PTS", "GmSc", "+/-"]

# avoid the first header row
rows = soup.findAll('tr')[25:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]

stats = pd.DataFrame(player_stats, columns = headers)
display(stats)
```
